```python
#Find Correlations
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])
correlation = numeric_df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation, annot=True)
plt.show()
```

```python
#Find most important features for grand_total
correlation['grand_total'].sort_values(ascending=False)
```

```python
#Scatter Plot of Important Features
plt.scatter(df['qty_ordered'], df['grand_total'])
plt.xlabel("Quantity Ordered")
plt.ylabel("Grand Total")
plt.show()
```

```python
#Handling outliers
import numpy as np

Q1 = df['grand_total'].quantile(0.25)
Q3 = df['grand_total'].quantile(0.75)

IQR = Q3 - Q1

df = df[(df['grand_total'] >= Q1 - 1.5*IQR) & (df['grand_total'] <= Q3 + 1.5*IQR)]
```

Q:
What is the best-selling category?
A:
The best-selling category is the one with the highest number of orders in the dataset. By analyzing the frequency of the category_name_1 column, we can determine which category has the most sales.
The category with the highest count represents the most popular products among customers

```python
df['category_name_1'].value_counts()
```

```python
df['category_name_1'].value_counts().plot(kind='bar')
plt.title("Best Selling Category")
plt.show()
```

```python
#Find a correlation between payment method and order status
df['payment_method'].value_counts().plot(kind='bar')
plt.title("Payment Method Frequency")
plt.show()
```

```python
df['status'].value_counts().plot(kind='bar')
plt.title("Order Status Frequency")
plt.show()
```

```python
pd.crosstab(df['payment_method'], df['status'])
```

```python
#Correlation Between Order Date & Category
df['created_at'] = pd.to_datetime(df['created_at'])

df.groupby(df['created_at'].dt.month)['category_name_1'].value_counts()
```
